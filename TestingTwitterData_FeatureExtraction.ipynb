# Feature Extraction of DateTime

````python
import numpy as np
import matplotlib.pyplot as plt

climate_disaster.info()
````

````python
climate_disaster['created_at'].describe()
climate_disaster['created_at'].head()
````

````python
# convert date object strings to datetimes using pandas
climate_disaster['created_at'] = pd.to_datetime(climate_disaster['created_at'],format='%Y-%m-%dT%H:%M:%SZ', 
 errors = 'coerce')
````

````python
# We now have date time as the dtype of the column
climate_disaster['created_at'].head()
````

````python
# Find the created at year 
climate_disaster['created_at_year'] = climate_disaster['created_at'].dt.year
````

````python
climate_disaster[['created_at_year']].head()
````

````python
# Find the created at Month 
climate_disaster['created_at_month'] = climate_disaster['created_at'].dt.month
````

````python
climate_disaster[['created_at_month']].head()
````

````python
# Find the created at day 
climate_disaster['created_at_day'] = climate_disaster['created_at'].dt.day
````

````python
climate_disaster[['created_at_day']].head()
````

````python
# Find the day of the week for each created at date
climate_disaster['created_at_dayofweek'] = climate_disaster['created_at'].dt.dayofweek
````

````python
climate_disaster[['created_at_dayofweek']].head()
````
